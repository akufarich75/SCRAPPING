## Retrive Tweets from Key Word

```python
import re
```

```python
search_words='Iqbaal'
date_since='2020-09-04'
```

```python
new_search = search_words + " -filter:retweets"
tweets = tweepy.Cursor(api.search, q=new_search, lang='id', since=date_since,
                       tweet_mode='extended').items(300)
```

```python
items = []
for tweet in tweets:
    item = []
    item.append(tweet.user.screen_name)
    item.append(tweet.user.location)
    item.append(' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet.full_text).split()))
#     item.append(tweet.full_text)
    items.append(item)
```

```python
items
```

### Make DataFrame

```python
import pandas as pd
```

```python
data = pd.DataFrame(items, columns=['nama', 'lokasi', 'tweet'])
```
